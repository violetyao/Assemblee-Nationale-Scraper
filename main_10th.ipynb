```python
import requests
import os.path
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
target = 'http://www.assemblee-nationale.fr/10/documents/index-dossier.asp'
req = requests.get(url=target)
html = req.text
```

```python
with open('directory.txt', 'w') as file:
    file.write(html)
```

```python
urls = []
with open('directory.txt') as file:
    for line in file:
        start = line.find("/10/dos")
        end = line.find(".asp")
        if (start !=-1) and (end !=-1):
            urls.append("http://www.assemblee-nationale.fr"+line[start:end]+".asp")
```

```python
print(urls)
```

```python
save_path = './manifesto10th/'
for i, url in enumerate(urls):
    print(i, "doing" , url)
    text = requests.get(url=url).text
    with open(os.path.join(save_path, str(i)+ '.txt'), 'w') as file:
        file.write(text) 
```

```python
save_path = './manifesto10th/'
infos = []

for i in range(200):
    info = []
    contents= ""
    title = ""
    name = ""
    print(i , "first_step================================================================")
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try: 
            page_soup = bs(file,"html.parser")
#             unknown = page_soup.find_all("meta",content="")
#             print(unknown)
            content = page_soup.find_all(align="left")[1]
            content_text = page_soup.find_all(align="left")[1].text
            print("content ============",content)
            print("content_text =========", content_text)
#             ind_1 = content_text.split("\n")[0].find(content.find_all("a")[-2].text)+len(content.find_all("a")[-2].text)+2
#             ind_1 = content_text.split("\n")[0].find_all(" ")
# [m.start() for m in re.finditer('test', 'test test test test')][-2]
            ind_1 = [m.start() for m in re.finditer('  ', content_text.split("\n")[0])][-2]
#             ind_1 = content_text.split("\n")[0].find(content.find_all("  ")[-2].text)+len(content.find_all("  ")[-2].text)+2
            ind_2 = content_text.split("\n")[0].find("n")-2
            print("ind_2========", content.find_all("a")[-2].text)
            contents += content_text.split("\n")[0][ind_1:ind_2]
            print("content_extracted ============",content_text.split("\n")[0][ind_1:ind_2])
        except:
            print('go into except')
            pass
    
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try: 
            page_soup = bs(file,"html.parser")
            n_line = page_soup.find_all(target="_blank")
            print("n_line======", len(n_line),n_line)
            name += n_line[0].text
            for j in range(1,len(n_line)):
                n = n_line[j].text
                if n and (n not in name): #add multiple names and prevent repeats
                    name += ", "
                    name += n
            
        except:
            print('name go into except')
            pass
        
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try:
            page_soup = bs(file,"html.parser")
            t = page_soup.title.string[len('Assemble nationale -'):]
            title += t            
        except:
            pass

    info = [title, name, contents]    
    infos.append(info)
```

```python
df = pd.DataFrame(infos, columns=['Title', 'Name','content'])
df
```
