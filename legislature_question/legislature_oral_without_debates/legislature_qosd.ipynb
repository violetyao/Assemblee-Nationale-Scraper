```python
import requests
import os.path
import pandas as pd
from bs4 import BeautifulSoup as bs
```

end : http://questions.assemblee-nationale.fr/q15/15-630QOSD.htm

start : http://questions.assemblee-nationale.fr/q15/15-1QOSD.htm

```python
urls = []
for i in range(1,630):
    url = "http://questions.assemblee-nationale.fr/q15/15-"+ str(i) + "QOSD.htm"
    urls.append(url)
```

```python
save_path = './questions/'
for i, url in enumerate(urls,1):
    print(i, "doing" , url)
    text = requests.get(url=url).text
    with open(os.path.join(save_path, str(i)+ '.txt'), 'w') as file:
        file.write(text) 
```

```python
save_path = './questions/'
infos = []

# first just get question content and name of the politician

for i in range(1,630):
    info = []
    index = ""
    name = ""
    ministere_interrogem = ""
    question = "" 
    date_publication = ""
    print(i , "first_step================================================================")
    with open(os.path.join(save_path, str(i)+ '.txt'), 'r') as file:
        try: 
            # add index
            index += str(i)
        
            page_soup = bs(file,"html.parser")
            
            # get name
            name_raw = page_soup.find("div",{"id":"question_col80"})
            name += name_raw.find('a').text
#             print(name)

            # get ministere_interroge
            minist_raw = page_soup.find("div",{"class":"question_col50"}).text
            startkey = "> "
            start = minist_raw.find(startkey) + 1
            ministere_interrogem += minist_raw[start:].strip()
            
            # get date_publication
            date_raw = page_soup.find("div",{"class":"question_publish_date"})
            date = date_raw.find('span').text
            date_publication += date
#             print(date)
            
            # get question
            question_raw = page_soup.find("div", {"class": "question"})
            question = question_raw.find('p').text
            # get rid of '\n'
            question = question.strip()
#             print(question)
                        
        except:
            print("go into except")
            pass
#     info = [question] 
    info = [index,name,ministere_interrogem,date_publication, question]    
    infos.append(info)
```

```python
df = pd.DataFrame(infos, columns=['index','nom_depute','ministere_interrogem',"date_publication",'question'])
df
```

```python
df.to_csv('legislature_qosd.csv')
```
